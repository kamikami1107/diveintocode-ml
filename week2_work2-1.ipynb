# ダミーデータの作成

### 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\left[
\begin{array}{ccc}
  1.0 & 0.8\\
  0.8 & 1.0
\end{array}
\right]
$$

多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。
作成したndarrayのshapeは(500, 2)になります。

```python
np.random.seed(0)
```

```python
import numpy as np
mean = ([-3, 0])
cov = ([[1.0, 0.8], [0.8, 1.0]])
size = 500
data1 = np.random.multivariate_normal(mean, cov, size)
print(data1.shape)
```

```python
print(data1)
```

### 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibの
#### plt.scatter()
を使うことで描けます。

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.title("scatter", size=20, weight='heavy')#大きさ20,太文字
plt.xlabel("x1", size=18)
plt.ylabel("x2", size=18)
plt.tick_params(labelsize=12)#軸のラベルサイズ12(size=だと数字を示す棒の長さのカスタマイズ)
plt.scatter(data1[:, 0:1], data1[:, 1:2])
plt.show()
```

### 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。

ヒストグラムは
#### plt.hist()
を使うことで描けます。

```python
data1_min = data1.min()
data1_max = data1.max()

plt.title("histogram_x1", size=20, weight='light', c='orange')#文字サイズ20,細文字,色をオレンジ
plt.xlabel("x1", size=18)
plt.ylabel("distribution", size=18)
plt.tick_params(labelsize=12)#軸のラベルサイズ12
plt.hist(data1[:, 0], bins=50)
plt.xlim([data1_min, data1_max])
plt.show()              

plt.title("histogram_x2", size=20, weight='medium', c='green')#文字の太さ通常
plt.xlabel("x2", size=18)
plt.ylabel("distribution", size=18)
plt.tick_params(labelsize=12)
plt.hist(data1[: ,1], bins=50)
plt.xlim([data1_min, data1_max])
plt.show()
```

### 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\left[
\begin{array}{ccc}
  1.0 & 0.8\\
  0.8 & 1.0
\end{array}
\right]
$$

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。

```python
mean = ([0, -3])
cov = ([[1.0, 0.8], [0.8, 1.0]])
size = 500
data2 = np.random.multivariate_normal(mean, cov, size)
print(data2.shape)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.title("scatter2", size=20, weight='heavy')#文字サイズを20,太文字
plt.xlabel("x1", size=18, c='red')#文字サイズを18, 色を赤色
plt.ylabel("x2", size=18, c='blue')##文字サイズを18, 色を青色
plt.tick_params(labelsize=12)#軸のラベルサイズを12
plt.scatter(data1[: ,0], data1[: ,1], label="0", c='blue')#ラベル名を追加、色を青色
plt.scatter(data2[: ,0], data2[: ,1], label="1", c='red')#ラベル名を追加、色を赤色
plt.legend()#凡例を表示
plt.show()
```

### 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate( )やnp.vstack( )を使うことで行えます。

→(1000, 2)の配列を作るということは、(500, 2)の配列を縦に2つ連結させるということである。

- np.concatenate( )

```python
data = np.concatenate([data1, data2], axis=0)
print(data.shape)
```

- np.vstack( )

```python
data = np.vstack([data1, data2])
print(data.shape)
```

**【質問】グラフを出してみたら、散布図が２個出てきました。問１のデータは(500,2)で散布図が一つなのに、結合データは(1000, 2)で散布図が２個表示されてしまうのはなぜでしょか？結合して配列は一つでも、元は２つの配列だからでしょうか？？**

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.title("scatter3", size=20, weight='heavy')
plt.xlabel("x1", size=18, c='red')
plt.ylabel("x2", size=18, c='blue')
plt.tick_params(labelsize=12)
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

### 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

```python
#(500, 1)の、0と1の配列をそれぞれ作成し、それを縦に連結
#その後、問１と問４の配列に横付けする

sample_zeros = np.zeros(500)
sample_zero1 = sample_zeros.reshape(500, 1)
#print(sample_zeros1)
#print(sample_zeros1.shape)

sample_ones = np.ones(500)
sample_ones1 = sample_ones.reshape(500, 1)
#print(sample_ones1)
#print(sample_ones1.shape)

label = np.concatenate([sample_zero1, sample_ones1], axis=0)#縦に連結
data_with_label = np.concatenate([data, label], axis=1)#横に連結
print(data_with_label)
print(data_with_label.shape)
```
