# 曽呂利新左衛門問題

#### 【逸話】
秀吉から褒美を下される際、何を希望するか尋ねられた新左衛門は、今日は米1粒、翌日には倍の2粒、その翌日には更に倍の4粒と、日ごとに倍の量の米を100日間もらう事を希望した。米粒なら大した事はないと思った秀吉は簡単に承諾したが、日ごとに倍ずつ増やして行くと100日後には膨大な量になる事に途中で気づき、他の褒美に変えてもらった。

## 【問題１】100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

- **①100日目までにもらう米粒が、累計何粒になるかを算出する**
- **②「その日にもらう米粒の数」と「その日までに累計で貰う米の数」をグラフ化する**  

````python
# ①100日目までにもらう米粒が、累計何粒になるかを算出する
days = 100

list_n_grains = []#ある日にもらう米の数のリスト
list_total_grains = []#ある日までに累計でもらう米の数のリスト

list_n_grains.append(1)
list_total_grains.append(1)

for i in range(1, days):
    list_n_grains.append(list_n_grains[-1]*2)
    list_total_grains.append(sum(list_n_grains))
    

print("{}日目にもらう米粒の数は{}粒です。".format(days, list_n_grains[days-1]))
print("{}日目までに累計で貰う米の数は{}粒です。".format(days, list_total_grains[days-1]))
#print(list_n_grains)
#print(list_total_grains)
#print(len(list_n_grains))
#print(len(list_total_grains))
````

- ②「その日にもらう米粒の数(in a day)」と「その日までに累計で貰う米の数(total)」をグラフ化する

````python
from matplotlib import pyplot as plt
%matplotlib inline
plt.title("rice table of Shinzaemon", size=20, weight='heavy')
plt.xlabel("days", size=18)
plt.ylabel("quantity of rice", size=18)
plt.tick_params(labelsize=12)
plt.plot(list_n_grains[1:days+1], label="in a day")
plt.plot(list_total_grains[1:days+1], label="total")
plt.legend()#2つの凡例表示
plt.show()
````
